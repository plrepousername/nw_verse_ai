```python
import zipfile
from io import BytesIO

jwpub_path = "nwt_X.jwpub"

with zipfile.ZipFile(jwpub_path, 'r') as z:
    # contents laden
    contents_bytes = z.read("contents")

    # inneres ZIP öffnen
    inner_zip = zipfile.ZipFile(BytesIO(contents_bytes))

    print("Dateien im inneren ZIP:")
    for f in inner_zip.namelist():
        print("-", f)

    # DB finden
    db_name = [f for f in inner_zip.namelist() if f.endswith(".db")][0]
    print("\nGefundene DB:", db_name)

    db_bytes = inner_zip.read(db_name)
```

```python
import tempfile
import sqlite3

tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".db")
tmp.write(db_bytes)
tmp.close()

conn = sqlite3.connect(tmp.name)
```

```python
#!/usr/bin/env python
# coding: utf-8

import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup
import re
import json


# -------------------------
# EPUB → TEXT
# -------------------------
def extract_text_from_epub(epub_path):
    book = epub.read_epub(epub_path)
    full_text = []

    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            soup = BeautifulSoup(item.get_content(), 'html.parser')

            # 🔥 wichtig für Struktur
            text = soup.get_text(separator="\n")

            full_text.append(text)

    return '\n'.join(full_text)


# -------------------------
# ZEILEN NORMALISIEREN (FIX)
# -------------------------
def normalize_lines(text):
    lines = text.split("\n")
    cleaned = []

    buffer = ""

    for line in lines:
        line = line.strip()

        if not line:
            continue

        # neue Bibelstelle erkannt → buffer flushen
        if re.match(r'^[\d\w.\s]+ \d+:\d+', line):
            if buffer:
                cleaned.append(buffer)
                buffer = ""
            cleaned.append(line)
        else:
            buffer += " " + line

    if buffer:
        cleaned.append(buffer)

    return cleaned


# -------------------------
# PARSER (UNVERÄNDERT)
# -------------------------
def parse_bible_text_file_from_lines(lines):
    bible_data = {}
    current_book = None
    current_chapter_num = None
    current_chapter_content = []

    print("Parsing gestartet...")

    for stripped_line in lines:
        stripped_line = stripped_line.strip()
        if not stripped_line:
            continue

        # 1. Header-Match
        header_match = re.match(r'^\s*([\w.\s]+?)\s*(?:\([^)]*\))?\s+(\d+)\s*:\s*\d+', stripped_line)
        if header_match:
            if current_book and current_chapter_num is not None and current_chapter_content:
                parse_chapter_block(bible_data, current_book, current_chapter_num, current_chapter_content)

            current_book = re.sub(r'\s+', ' ', header_match.group(1).strip())
            current_chapter_num = int(header_match.group(2))
            current_chapter_content = []
            continue

        # 2. Ein-Kapitel-Bücher
        sorted_books = sorted(one_chapter_books, key=len, reverse=True)
        buchnamen_regex = '|'.join(re.escape(name) for name in sorted_books)

        pattern = rf'^({buchnamen_regex})\s+1\b'
        stripped_line = stripped_line.replace("\xa0"," ")
        one_chapter_match = re.match(pattern, stripped_line)

        if one_chapter_match:
            if current_book and current_chapter_num is not None and current_chapter_content:
                parse_chapter_block(bible_data, current_book, current_chapter_num, current_chapter_content)

            current_book = one_chapter_match.group(1)
            current_chapter_num = 1
            current_chapter_content = []
            print(f"📘 Ein-Kapitel-Buch erkannt: {current_book}")
            continue

        # 3. Content sammeln
        if current_book and current_chapter_num is not None:
            if not stripped_line.startswith('^') and not stripped_line.isupper():
                current_chapter_content.append(stripped_line)

    # letzter Block
    if current_book and current_chapter_num is not None and current_chapter_content:
        parse_chapter_block(bible_data, current_book, current_chapter_num, current_chapter_content)

    print("Parsing fertig.")
    return bible_data


# -------------------------
# CHAPTER PARSER (UNVERÄNDERT)
# -------------------------
def parse_chapter_block(bible_data, book_name, chapter_num, content_lines):
    book_name = book_name.replace("\xa0"," ")

    if book_name not in bible_data:
        bible_data[book_name] = {}

    verse_dict = {}
    full_content = " ".join(content_lines)
    full_content = re.sub(r'\s+', ' ', full_content).strip()

    pattern_to_remove = r'^\s*' + str(chapter_num) + r'\s+'
    if re.match(pattern_to_remove, full_content):
        full_content = re.sub(pattern_to_remove, '', full_content, 1)

    all_matches = list(re.finditer(r'\b(\d+)\b', full_content))

    actual_verse_markers = []

    if book_name == JOHANNES_BOOKNAME and chapter_num == 8:
        last_real_verse_num = 11
    else:
        last_real_verse_num = 1

    for match in all_matches:
        current_num = int(match.group(1))
        if current_num == last_real_verse_num + 1:
            actual_verse_markers.append(match)
            last_real_verse_num = current_num

    if not actual_verse_markers:
        if full_content:
            verse_dict[1] = full_content
    else:
        first_marker_start_pos = actual_verse_markers[0].start()
        verse_dict[1] = full_content[:first_marker_start_pos].strip()

        for i, marker in enumerate(actual_verse_markers):
            verse_num = int(marker.group(1))
            text_start_pos = marker.end()
            text_end_pos = actual_verse_markers[i+1].start() if i + 1 < len(actual_verse_markers) else len(full_content)

            verse_text = full_content[text_start_pos:text_end_pos].strip()
            verse_dict[verse_num] = verse_text

    if verse_dict:
        max_verse = max(verse_dict.keys())
        last_verse_text = verse_dict[max_verse]

        match = re.search(r'\b1\b\s+[A-ZÄÖÜa-zäöü]', last_verse_text)
        if match:
            cutoff_pos = match.start()
            verse_dict[max_verse] = last_verse_text[:cutoff_pos].strip()

    bible_data[book_name][chapter_num] = verse_dict


# -------------------------
# ENTRY FUNCTION
# -------------------------
def epub_to_bible_json(epub_path, output_json):
    print("📖 EPUB lesen...")
    text = extract_text_from_epub(epub_path)

    print("🧹 Zeilen normalisieren...")
    lines = normalize_lines(text)

    print("🔍 Parser starten...")
    bible_data = parse_bible_text_file_from_lines(lines)

    print("💾 JSON speichern...")
    with open(output_json, "w", encoding="utf-8") as f:
        json.dump(bible_data, f, ensure_ascii=False, indent=2)

    print("✅ Fertig!")


# -------------------------
# CONFIG
# -------------------------
one_chapter_books = {"Obadja", "Philemon", "2. Johannes", "3. Johannes", "Judas"}
JOHANNES_BOOKNAME = "Johannes"


# -------------------------
# RUN
# -------------------------
if __name__ == "__main__":
    epub_to_bible_json("nwt_X.epub", "bibel.json")
```
